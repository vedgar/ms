````python
class PrirodniBrojevi:
    def __iter__(self):
        brojač = 0
        while True:
            yield brojač
            brojač += 1
````

````python
omega = PrirodniBrojevi()
for n in omega:
    print(n)
    if n > 10: breakvars(A)
````

````python
def Fibonacci():
    prvi, drugi = 0, 1
    while True:
        yield prvi
        prvi, drugi = drugi, prvi + drugi
````

````python
any(str(broj).endswith('0000') for broj in Fibonacci())
````

````python
class Velik:
    def __init__(self, broj):
        self.vrijednost = broj

    def __bool__(self):
        return self.vrijednost > 10
````
